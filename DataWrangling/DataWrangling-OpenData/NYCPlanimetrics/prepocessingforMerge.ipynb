```python
import pandas as pd
import os
import rtree
import osmnx
```

```python
final_dt = pd.DataFrame(columns = ['segid','id'])
final_dt
```

```python

for fc in os.listdir():
    if fc.startswith('joined'):
        dt_tmp = pd.read_csv(fc)
        final_dt = pd.merge(final_dt,dt_tmp,left_on='segid',right_on='segid',how = 'outer')
final_dt = final_dt.drop(columns = 'id').fillna(0)
```

```python
final_dt['new_sidewalk'] = final_dt.sidewalk.apply(lambda x: 2 if x >=2 else x)
```

```python
final_dt.drop( 'sidewalk',inplace = True,axis = 1)
final_dt.head()
```

```python
final_dt.rename(columns={'new_sidewalk':'sidewalk'},inplace = True)
final_dt
```

```python
final_dt['playsprt'] = final_dt['playsprt_baseball'] + final_dt['playsprt_football'] + final_dt['playsprt_swim']
```

```python
#apply on binary feature
```

```python
def binaryfeature(feature,dt = final_dt):
    dt['tmp'] = dt[feature].apply(lambda x: 1 if x>=1 else x)
    dt[feature] = dt['tmp']
    dt.drop('tmp', inplace=True, axis = 1)
    return dt
```

```python
featurelist = ['beach','pubgardn', 'othrsdwk','plazasq','median','surfpkng','pbspothr','abndbldg','playsprt','parkplay']
for feature in featurelist:
    final_dt = binaryfeature(feature,dt = final_dt)
```

```python
final_dt.drop(columns = ['playsprt_baseball','playsprt_football','playsprt_swim'],inplace=True)
```
